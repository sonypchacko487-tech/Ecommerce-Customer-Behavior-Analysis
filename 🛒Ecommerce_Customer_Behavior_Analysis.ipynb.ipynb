 🧹 Data Cleaning & Pre-processing of E-Commerce Customer Dataset

### 🎯 <u>Objective</u>
The objective of this phase is to clean and prepare the dataset by handling missing values, duplicates, inconsistencies, and incorrect data formats to ensure accurate and reliable analysis.

## 🔍 Create a Copy of the Original Dataset

```python
df_clean = df.copy()
```

## 🔤 Standardize Column Names 

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
```

## 🕳️ Handle Missing Values 

```python
df_clean.isnull().sum()
```

## ✅ Handle Missing Value (Numerical Columns)

```python
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())
df_clean['Average_Order_Value'] = df_clean['Average_Order_Value'].fillna(
    df_clean['Average_Order_Value'].mean()
)
df_clean['Session_Duration_Avg'] = df_clean['Session_Duration_Avg'].fillna(
    df_clean['Session_Duration_Avg'].mean()
)
df_clean['Pages_Per_Session'] = df_clean['Pages_Per_Session'].fillna(
    df_clean['Pages_Per_Session'].median()
)
```

## ✅ Handle  Missing Values (Categorical Columns)

```python
df_clean['Gender'] = df_clean['Gender'].fillna(
    df_clean['Gender'].mode()[0]
)
df_clean['Country'] = df_clean['Country'].fillna(
    df_clean['Country'].mode()[0]
)
df_clean['City'] = df_clean['City'].fillna(
    df_clean['City'].mode()[0]
)
df_clean['Signup_Quarter'] = df_clean['Signup_Quarter'].fillna(
    df_clean['Signup_Quarter'].mode()[0]
)
```

**Insight:** Missing values were primarily found in numerical columns and were handled using median imputation to avoid distortion caused by outliers.

## ♻️ Remove Duplicate Records 

```python
df_clean.drop_duplicates(inplace=True)
```

 ## 🧹 Standardize Categorical Text Values 

```python
df_clean['Gender'] = df_clean['Gender'].str.strip().str.capitalize()
df_clean['Country'] = df_clean['Country'].str.strip().str.title()
df_clean['City'] = df_clean['City'].str.strip().str.title()
```

## 🧪 Data Type Validation & Correction 

```python
df_clean['Churned'] = df_clean['Churned'].astype(int)

df_clean['Mobile_App_Usage'] = (
    df_clean['Mobile_App_Usage']
    .fillna(0)
    .astype(int)
)
```

## 📉 Outlier Detection & Treatment 

### 🔍 IQR Method on Average Order Value

```python
Q1 = df_clean['Average_Order_Value'].quantile(0.25)
Q3 = df_clean['Average_Order_Value'].quantile(0.75)
IQR = Q3 - Q1

df_clean = df_clean[
    (df_clean['Average_Order_Value'] >= Q1 - 1.5 * IQR) &
    (df_clean['Average_Order_Value'] <= Q3 + 1.5 * IQR)
]
```

## ✅ Final Data Quality Check 

```python
df_clean.info()
df_clean.isnull().sum()
df_clean.shape
```

##  💾 Save the Cleaned Dataset

```python
df_clean.to_csv("cleaned_ecommerce_customer_behaviour_dataset.csv", index=False)
```
